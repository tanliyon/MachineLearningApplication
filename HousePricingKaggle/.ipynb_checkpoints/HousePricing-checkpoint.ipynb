```python
## Dropping similar features

train = train.drop('TotalBsmtSF', axis = 1)
train = train.drop('GarageArea', axis = 1)
```

```python
## Feature Engineering
trainingDataRows = train.shape[0]
salePriceTraining = train.SalePriceLog.values

allData = pd.concat((train, test)).reset_index(drop=True)
allData.drop(['SalePrice', 'SalePriceLog'], axis=1, inplace=True)
print(allData.shape)
allData.head()

# First, we combine the data from both training set and test set and drop the SalePrice column.
# This is so that we can focus on the features only and inspect for missing features
```

```python
## Check for missing variables

allDataNaRatio = (allData.isnull().sum() / len(allData)) * 100
allDataNaRatio = allDataNaRatio.drop(allDataNaRatio[allDataNaRatio == 0].index).sort_values(ascending=False)
ratioTable = pd.DataFrame(allDataNaRatio)
ratioTable

# From the data, we can see that a lot of PoolQC is N/A, and according to the data description, it means that that house doesn't have a pool.
# The same applies for a lot of features, hence we can fill them all with 0
```

```python
## Filling in trivial missing data
for col in ('PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 'GarageQual', 'GarageCond', 'GarageFinish', 'GarageType', 'BsmtExposure', 'BsmtCond', 'BsmtQual', 'BsmtFinType2', 'BsmtFinType1', 'MasVnrType' ):
    allData[col] = allData[col].fillna('None')

for col in ('GarageYrBlt', 'MasVnrArea', 'BsmtFullBath', 'BsmtHalfBath', 'BsmtUnfSF', 'TotalBsmtSF', 'GarageArea', 'GarageCars', 'BsmtFinSF2', 'BsmtFinSF1'):
    allData[col] = allData[col].fillna(0)
    
# The N/A in these datas = none / 0 according to the feature description
```

```python
## Verify that the features has been dropped
allDataNaRatio = (allData.isnull().sum() / len(allData)) * 100
allDataNaRatio = allDataNaRatio.drop(allDataNaRatio[allDataNaRatio == 0].index).sort_values(ascending=False)
ratioTable = pd.DataFrame(allDataNaRatio)
ratioTable

# Hurray! Now we are left with only 9 features to predict
```

```python
## Filling in non-trivial missing data

allData['LotFrontage'] = allData.groupby("Neighborhood")["LotFrontage"].transform(
    lambda x: x.fillna(x.median()))
# We can estimate the lotfrontage of a house based on the neighborhood

allData['MSZoning'] = allData['MSZoning'].fillna(allData['MSZoning'].mode()[0])
# We will fill in MSZoning with the mode, which is RL

allData['Functional'] = allData['Functional'].fillna('Typ')
# This is following the feature description

for col in ('Utilities', 'SaleType', 'KitchenQual', 'Exterior2nd', 'Exterior1st', 'Electrical'):
    allData[col] = allData[col].fillna(allData[col].mode()[0])
# For these features, there are very few missing data and many of the same data. Hence we will fill it with mode.
```

```python
## Last check for missing data
allDataNaRatio = (allData.isnull().sum() / len(allData)) * 100
allDataNaRatio = allDataNaRatio.drop(allDataNaRatio[allDataNaRatio == 0].index).sort_values(ascending=False)
ratioTable = pd.DataFrame(allDataNaRatio)
ratioTable

# No more missing datas!
```

```python
## Label encoding for all categorical data

allData['MSSubClass'] = allData['MSSubClass'].astype(str)
allData['OverallCond'] = allData['OverallCond'].astype(str)
allData['YrSold'] = allData['YrSold'].astype(str)
allData['MoSold'] = allData['MoSold'].astype(str)

# First, we transform all the numerical data that is actually categorical data to categorical data.

from sklearn.preprocessing import LabelEncoder

cols = ('FireplaceQu', 'BsmtQual', 'BsmtCond', 'GarageQual', 'GarageCond', 
        'ExterQual', 'ExterCond','HeatingQC', 'PoolQC', 'KitchenQual', 'BsmtFinType1', 
        'BsmtFinType2', 'Functional', 'Fence', 'BsmtExposure', 'GarageFinish', 'LandSlope',
        'LotShape', 'PavedDrive', 'Street', 'Alley', 'SaleType', 'CentralAir','Utilities', 'MSSubClass', 'OverallCond', 
        'YrSold', 'MoSold', 'BldgType')

for c in cols:
    lbl = LabelEncoder() 
    lbl.fit(list(allData[c].values)) 
    allData[c] = lbl.transform(list(allData[c].values))

allData.head(20)

# Then we convert all the categorical data to labels, so it will turn all the wording to numerical values
# Label encoding is neccesary since the Scikit algorithms will be expecting the features to be in numbers instead of words
```

```python
## Convert to dummy variable
allData = pd.get_dummies(allData)
allData.head()

# Dummy variable is used to convert all the categorical data to binary, so that linear regression will not be confused by
# the numerical implication of label encoding. 
```

```python
## Splitting test set from training test
train = allData[:trainingDataRows]
test = allData[trainingDataRows:]

print(train.shape)
print(test.shape)
```

```python
## Importing all the libraries
from sklearn.linear_model import ElasticNet, Lasso,  BayesianRidge, LassoLarsIC
from sklearn.ensemble import RandomForestRegressor,  GradientBoostingRegressor
from sklearn.kernel_ridge import KernelRidge
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.base import BaseEstimator, TransformerMixin, RegressorMixin, clone
from sklearn.model_selection import KFold, cross_val_score, train_test_split
from sklearn.metrics import mean_squared_error
import xgboost as xgb
import lightgbm as lgb
```

```python
## Cross validation error
def crossValidationError (model):
    kf = KFold(5, shuffle=True).get_n_splits(train.values) # To shuffle the dataset and split into 5 folds
    rmse= np.sqrt(-cross_val_score(model, train.values, salePriceTraining, scoring="neg_mean_squared_error", cv = kf))
    return(rmse)

# The first step of splitting into 5 folds is needed because the cross_val_score itself does not shuffle the dataset.
# For cross validation with scoring = neg_mean_squared_error, a negative will need to be added in front to ensure that the term
# under the sqrt is positive.

# Cross validation works by splitting the test set into x folds. Then, we will train on x-1 folds, and test it on x.
# Then, the process repeats by this time, we use folds 2-x to train and test it on fold 1. 
# The main purpose of cross validation is to ensure that our model don't overfit and can generalize
```

```python
## Defining models

# Making pipeline for lasso and Enet ensures that our data passes through RobustScaler before passing into lasso.
# RobustScaler helps normalize the data by taking each point, minus the median and divide it by the interquatile range.
# As a result, the outliers have as big as an impact.
lasso = make_pipeline(RobustScaler(), Lasso(alpha =0.0005))

# Lasso works like a normal linear regression but the squared error will have a penalizing term at the end
# with alpha controlling how harsh to penalize the model. It does so by making sure that the adjustment of the weights
# are kept to as small as possible, so that the polynomial term will turn out smooth and not overfit.

ENet = make_pipeline(RobustScaler(), ElasticNet(alpha=0.0005, l1_ratio=.9))

# ENet works similar to a lasso, but it gives a bit more room for the weights to wiggle around.

GBoost = GradientBoostingRegressor(n_estimators=3000, learning_rate=0.05,
                                   max_depth=4, max_features='sqrt',
                                   min_samples_leaf=15, min_samples_split=12, 
                                   loss='huber')

XgBoost = xgb.XGBRegressor(colsample_bytree=0.4603, gamma=0.0468, 
                             learning_rate=0.05, max_depth=3, 
                             min_child_weight=1.7817, n_estimators=2200,
                             reg_alpha=0.4640, reg_lambda=0.8571,
                             subsample=0.5213, silent=1, nthread = -1)

Lgb = lgb.LGBMRegressor(objective='regression',num_leaves=5,
                              learning_rate=0.05, n_estimators=720,
                              max_bin = 55, bagging_fraction = 0.8,
                              bagging_freq = 5, feature_fraction = 0.2319,
                              feature_fraction_seed=9, bagging_seed=9,
                              min_data_in_leaf =6, min_sum_hessian_in_leaf = 11)

# GBoost, XgBoost and LGBM works almost the same way. It will construct a stronger learning from a weak learner.
# Once we have a model, we will notice that some of the data does not sit well with the model, hence, they will be selected to
# train another model. Model 2 will be different from model 1 since it will try to fit more towards the errors.
# The process will keep on iterating and at the end, the models will be added together and the average will be taken.
```

```python
## Lasso performance
crossValidationError(lasso).mean()
```

```python
## Enet performance
crossValidationError(ENet).mean()
```

```python
## GBoost performance
crossValidationError(GBoost).mean()
```

```python
## XgBoost performance
crossValidationError(XgBoost).mean()
```

```python
## Lgb performance
crossValidationError(Lgb).mean()
```

```python
## Inherit the properties from SKLearn and change the fit and predict method
## The whole part is still quite confusing, credit goes to https://www.kaggle.com/serigne/stacked-regressions-top-4-on-leaderboard

class StackingAveragedModels(BaseEstimator, RegressorMixin, TransformerMixin):
    def __init__(self, base_models, meta_model, n_folds=5):
        self.base_models = base_models
        self.meta_model = meta_model
        self.n_folds = n_folds
   
    def fit(self, X, y):
        self.base_models_ = [list() for x in self.base_models]
        self.meta_model_ = clone(self.meta_model)
        kfold = KFold(n_splits=self.n_folds, shuffle=True, random_state=156)
        
        out_of_fold_predictions = np.zeros((X.shape[0], len(self.base_models)))
        for i, model in enumerate(self.base_models):
            for train_index, holdout_index in kfold.split(X, y):
                instance = clone(model)
                self.base_models_[i].append(instance)
                instance.fit(X[train_index], y[train_index])
                y_pred = instance.predict(X[holdout_index])
                out_of_fold_predictions[holdout_index, i] = y_pred
                
        self.meta_model_.fit(out_of_fold_predictions, y)
        return self
   
    def predict(self, X):
        meta_features = np.column_stack([
            np.column_stack([model.predict(X) for model in base_models]).mean(axis=1)
            for base_models in self.base_models_ ])
        return self.meta_model_.predict(meta_features)
    
# The general working idea is to first split the training data into folds, then train the models on x-1 folds with the base models.
# After x iterations, all the folds will have been predicted on. 
# The output will then be feed into the metamodel as a new feature and it will predict the final output.
```

```python
## Defining the RMS Error function
def rmsle(y, y_pred):
    return np.sqrt(mean_squared_error(y, y_pred))
```

```python
## Checking for the improvement in stacking the models
stackedModels = StackingAveragedModels(base_models = (ENet, GBoost),
                                                 meta_model = lasso)
crossValidationError(stackedModels).mean()

# Just by stacking, the model improved by a lot!
```

```python
## Fitting the stacked models and converter the logged sale price back to normal value
stackedModels.fit(train.values, salePriceTraining)
stackedPrediction = stackedModels.predict(train.values)
stackedExpPrediction = np.expm1(stackedModels.predict(test.values))
rmsle(salePriceTraining, stackedPrediction)
```

```python
## Fitting the XgBoost model and converter the logged sale price back to normal value
XgBoost.fit(train.values, salePriceTraining)
XgBoostPrediction = XgBoost.predict(train.values)
XgBoostExpPrediction = np.expm1(XgBoost.predict(test.values))
print(rmsle(salePriceTraining, XgBoostPrediction))
```

```python
## Fitting the Lgb model and converter the logged sale price back to normal value
Lgb.fit(train.values, salePriceTraining)
LgbPrediction = Lgb.predict(train.values)
LgbExpPrediction = np.expm1(Lgb.predict(test.values))
print(rmsle(salePriceTraining, LgbPrediction))
```

```python
## Error calculation on training data
rmsle(salePriceTraining,stackedPrediction * 0.60 +
               XgBoostPrediction * 0.2 + LgbPrediction * 0.2)
```

```python
## Stacking the exponential prediction onto an output
finalSalePricePrediction = stackedExpPrediction * 0.70 + XgBoostExpPrediction * 0.15 + LgbExpPrediction * 0.15
```

```python
sub = pd.DataFrame()
sub['Id'] = testID
sub['SalePrice'] = finalSalePricePrediction
sub.to_csv('submission.csv',index=False)
```
